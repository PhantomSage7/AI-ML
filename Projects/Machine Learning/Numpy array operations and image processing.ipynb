## Shape of Data / Distribution 

```python
data = np.random.randint(1, 100, 1000)
# uniform distribution
plt.figure(figsize=(8, 3))
plt.hist(data, bins=30)
plt.show()
```

```python
height = np.random.normal(size=1000)
# standard normal distribution mean=0, std=1
plt.figure(figsize=(8, 3))
plt.hist(height, bins=30, ec='k', label="standard normal distribution")
plt.legend()
plt.show()
```

```python
height = np.random.normal(160, 15, 1000)
# standard normal distribution mean=0, std=1
plt.figure(figsize=(8, 3))
plt.hist(height, bins=30, ec='k', label="standard normal distribution")
plt.legend()
plt.show()
```

```python
new_data = np.append(height, np.random.uniform(50, 100, 50))
new_data = np.append(new_data, np.random.uniform(100, 130, 50))
```

```python
plt.figure(figsize=(10, 3)) # uniform, normal distribution, skweness in data
height = np.random.normal(160, 15, 1000)
plt.hist(new_data, bins=30, label="left skewed normal distribution", ec='k')
bar = new_data.mean()
mu = np.median(new_data)

plt.vlines(bar, 0, 140, label="mean", color="yellow")
plt.vlines(mu, 0, 140, label="median", color="red")
plt.legend()
plt.show()
```

```python
new_data = np.append(height, np.random.uniform(180, 250, 120))

plt.figure(figsize=(10, 3))

plt.hist(new_data, bins=30, label="right skewed normal distribution")
bar = new_data.mean()
mu = np.median(new_data)

plt.vlines(bar, 0, 140, label="mean", color="yellow")
plt.vlines(mu, 0, 140, label="median", color="red")
plt.legend()
plt.show()
```
